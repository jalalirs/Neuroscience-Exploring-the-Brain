## <font color="10D2D4">Level of Analysis</font>

### <font color="F95050">Molecular Neuroscience</font>
- The study of the brain at the most elementary level

### <font color="F95050">Cellular Neuroscience</font>
- How many types of neurons are there? how they differ?
- How do neurons influence other neurons?
- How do neurons become wired together during fetal development?
- How do neurons perform computation?

### <font color="F95050">System Neuroscience</font>
- The study of the complex circutes that perform a common function, such as vision or voluntary movement. 

### <font color="F95050">Behavioral Neuroscience</font>
- How do neural systems work together to produce integrated behaviors? 
- Where in the brain do "mind-altering" druns act, and what is the normal contribution of these systems to the regulation of mood and behaviors? 
- what neural systems account for gender-specific behaviors?
- Where are dreams created and what do theay reveal?

### <font color="F95050">Cognitive Neuroscience</font>
- The study of the higher levels of human mental activity, such as self-awareness, imagination, and language.
